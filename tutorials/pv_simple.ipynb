# Simple example of creating a PV model

```python
from consmodel import PV
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# create a simple PV model
pv = PV(lat=46.155768,
        lon=14.304951,
        alt=400,
        index=1,
        name="test",
        tz="Europe/Vienna")
```

```python
timeseries = pv.simulate(pv_size=14.,
                         year=2022,
                         freq="15min",
                         model="ineichen",
                         consider_cloud_cover=True)
```

```python
timeseries.plot()
plt.show()
```

```python
def show_week(df, week):
    df = df.loc[df.index.week == week]
    df.plot()
    plt.show()

show_week(timeseries, 1)
```
